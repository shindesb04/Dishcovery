Import the libraries

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

## loading model

```python
cnn = tf.keras.models.load_model('/content/trained_model.h5')
```

```python
#Performing the single prediction
import cv2
image_path = '/content/drive/MyDrive/Fruit_Vegetable_Recognition/test/corn/Image_7.jpg'
img = cv2.imread(image_path)
plt.imshow(img)
plt.title("Test Image")
plt.xticks([])
plt.yticks([])
plt.show
```

```python
# Testing the model
image = tf.keras.preprocessing.image.load_img(image_path,target_size=(64,64))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr]) # converting single image to batch
predictions = cnn.predict(input_arr)
```

```python
print(predictions)
```

```python
test_set = tf.keras.utils.image_dataset_from_directory(
    '/content/drive/MyDrive/Fruit_Vegetable_Recognition/test',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(64, 64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)
```

```python
result_index = np.where(predictions[0] == max(predictions[0]))
print(result_index[0])
```

```python
# Display the image
plt.imshow(img)
plt.yticks()
plt.xticks()
print("the Above image is predicted as",test_set.class_names[result_index[0][0]])
```
